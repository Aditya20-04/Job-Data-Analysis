```python
values=df[["experience_level","salary_in_usd"]].groupby(["experience_level"],as_index=False).mean().sort_values(by="salary_in_usd",ascending=False)
values=values["salary_in_usd"].values*100
labels=["EX","SE","MI",'EN']
a,fig=plt.subplots(figsize=(17,6))
plt.pie(x=values,labels=labels,autopct='%1.1f%%',explode=[0.04,0.04,0.04,0.04],shadow=True)
plt.axis('equal')
plt.style.use('dark_background')
plt.show()
```

```python
values=df[["experience_level","salary"]].groupby(["experience_level"],as_index=False).mean().sort_values(by="salary",ascending=False)
values=values["salary"].values*100
labels=["EX","SE","MI",'EN']
a,fig=plt.subplots(figsize=(17,6))
plt.pie(x=values,labels=labels,autopct='%1.1f%%',explode=[0.04,0.04,0.04,0.04],shadow=True)
plt.axis('equal')
plt.style.use('dark_background')
plt.show()
```

# How many jobs contain machine learning

```python
ml_jobs_df=df[df['job_title'].str.contains('Machine Learning')]
ml_jobs_df['job_title'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Machine Learning Jobs Distribution')
plt.show()
```

```python
ML_jobs_df=df[df['job_title'].str.contains('Machine Learning')]
ML_jobs_df['job_title'].value_counts().plot.bar(color='y',alpha=0.5)
plt.ylabel('count')
plt.show()
```
